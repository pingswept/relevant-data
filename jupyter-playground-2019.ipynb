```python
import plotly as py
import plotly.figure_factory as ff
import pandas as pd
py.offline.init_notebook_mode(connected=True)
```

```python
import cufflinks as cf
df = pd.DataFrame()
players = ['ray-allen', 'reggie-miller', 'steph-curry', 'james-harden', 'lebron-james', 'kyle-korver', 'klay-thompson']
for player in players:
    stats = pd.read_csv(player + "-totals-2019.csv").iloc[0:-1]
    df[player] = stats['3P'].cumsum()

py.offline.iplot(df.iplot(title='Cumulative Threes Made', xTitle='season', asFigure=True,kind='scatter', filename='cufflinks/cf-simple-line'),)
```

```python
import plotly.graph_objs as go

# Create random data with numpy
import numpy as np
import math

N = 500
x_values = np.linspace(-5, 1, N)

f1 = lambda x: x*np.exp(x)
trace1 = go.Scatter(
    x = x_values,
    y = f1(x_values)
)

f2 = lambda x: x**2*np.exp(x)
trace2 = go.Scatter(
    x = x_values,
    y = f2(x_values)
)

f3 = lambda x: x**3*np.exp(x)
trace3 = go.Scatter(
    x = x_values,
    y = f3(x_values)
)

py.offline.iplot([trace1, trace2, trace3], filename='basic-line')
```
